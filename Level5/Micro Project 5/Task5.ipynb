```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("D:/DS-DAY3/USA Housing Dataset.csv")
print("Column names in your dataset:")
print(df.columns.tolist())
print("\nSummary Statistics:")
print(df[['price', 'sqft_living']].describe())
df['price_per_sqft'] = df['price'] / df['sqft_living']
plt.figure(figsize=(8, 5))
plt.scatter(df['sqft_living'], df['price'], alpha=0.5, color='green')
plt.title('House Price vs. Living Square Footage')
plt.xlabel('Square Footage (Living)')
plt.ylabel('Price')
plt.grid(True)
print("\nUpdated DataFrame (first 5 rows):")
print(df[['price', 'sqft_living', 'price_per_sqft']].head())
plt.show()
```
