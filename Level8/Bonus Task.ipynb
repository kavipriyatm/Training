```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.pareto(a=2, size=1000)
plt.hist(data, bins=50, density=True, log=True)
plt.title("Pareto Distribution (log scale)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

sorted_data = np.sort(data)[::-1]
top_20 = sorted_data[:200]
total = data.sum()
top_sum = top_20.sum()
percentage = (top_sum / total) * 100

print(f"Top 20% of values ≈ {round(percentage)}% of total sum")
```
