```python
#task7
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
data_binomial = np.random.binomial(n=10, p=0.5, size=1000)
mean_binomial = data_binomial.mean()
plt.hist(data_binomial, bins=range(12), density=True, edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
print("Binomial Distribution Simulation")
print(f"Mean of Simulated Data: {mean_binomial:.2f}")
print("Matches binomial expectation (n × p = 5).\n")
```
