```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom

df = pd.read_csv('online_retail.csv', encoding='ISO-8859-1')
df = df.dropna(subset=['InvoiceNo', 'Quantity'])

invoice_totals = df.groupby('InvoiceNo')['Quantity'].sum().reset_index()
high_orders = invoice_totals[invoice_totals['Quantity'] > 50]

total_orders = len(invoice_totals)
high_order_days = len(high_orders)
p_high_order = high_order_days / total_orders

print(f"P(high order) = {p_high_order:.4f}")

x = range(0, 20)
pmf = binom.pmf(x, total_orders, p_high_order)

plt.bar(x, pmf)
plt.xlabel('Number of High Orders')
plt.ylabel('Probability')
plt.title('Binomial Distribution of High Orders')
plt.show()
```
